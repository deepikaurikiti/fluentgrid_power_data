```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('April Month data.xlsx')
df
```

```python
# List of columns to drop
columns_to_drop = ['kWhLead', 'IN', 'kWhLag']

# Drop the specified columns
df = df.drop(columns=columns_to_drop)

df
```

```python
# List of columns to check
columns_to_check = ['IY', 'IB', 'kVarh', 'IR']

# Check if all values in the specified columns are 0
all_zero = df[columns_to_check].eq(0).all()

# Print the result for each column
for column, is_all_zero in all_zero.items():
    print(f"Column '{column}' has all values as 0: {is_all_zero}")
```

```python
columns = ['kWh', 'VB', 'IY', 'IB', 'kVarh', 'IR', 'VR', 'VY']

statistics = df[columns].describe()

print(statistics)
```

```python
from sklearn.preprocessing import StandardScaler

# Specify the columns to be normalized
columns_to_normalize = ['kWh', 'VB', 'IY', 'IB', 'kVarh', 'IR', 'VR', 'VY']

# Normalize the column values using StandardScaler
scaler = StandardScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df[columns_to_normalize]), columns=columns_to_normalize)

print(df_normalized)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))
plt.plot(df['kWh'], label='kWh')
plt.plot(df['VB'], label='VB')
plt.plot(df['IY'], label='IY')
plt.plot(df['IB'], label='IB')
plt.plot(df['kVarh'], label='kVarh')
plt.plot(df['IR'], label='IR')
plt.plot(df['VR'], label='VR')
plt.plot(df['VY'], label='VY')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Line Chart of Energy Columns')
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df[['kWh', 'VB', 'IY', 'IB', 'kVarh', 'IR', 'VR', 'VY']])
plt.title('Scatter Plot Matrix of Energy Columns')
plt.show()
```

## Energy Consumption Analysis

```python
# Identify Peak Usage Periods

max_kWh = df['kWh'].max()
max_kWh_index = df['kWh'].idxmax()
print("Peak Usage Period:")
print("Max kWh Value:", max_kWh)
print("Index:", max_kWh_index)
```

```python
# Calculate Total Energy Consumption over a Specific Timeframe

start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-12-31')

total_kWh_consumption = df.loc[(df['rtimeid'] >= start_date) & (df['rtimeid'] <= end_date), 'kWh'].sum()
print("Total kWh Consumption:", total_kWh_consumption)
```

```python
# Compare Energy Usage between Different Periods or Locations

monthly_avg_kWh = df.groupby(df['rtimeid'].dt.month)['kWh'].mean()
print("Monthly Average kWh Consumption:")
print(monthly_avg_kWh)
```

## Power Factor Analysis

```python
# Calculate the Apparent Power (kVA)
df['Apparent_Power'] = np.sqrt(df['kWh']**2 + df['kVarh']**2)

# Calculate the Power Factor
df['Power_Factor'] = df['kWh'] / df['Apparent_Power']

# Analyze the Power Factor
# Descriptive Statistics
power_factor_stats = df['Power_Factor'].describe()
print("Power Factor Statistics:")
print(power_factor_stats)

# Histogram
plt.hist(df['Power_Factor'], bins=10)
plt.xlabel('Power Factor')
plt.ylabel('Frequency')
plt.title('Power Factor Distribution')
plt.show()
```

## Correlation Analysis

```python
# Calculate correlation matrix
correlation_matrix = df[['kWh', 'VB', 'IY', 'IB', 'kVarh', 'IR', 'VR', 'VY']].corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix')
plt.show()
```
